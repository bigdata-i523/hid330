```python
# monthly average AQI value, for the year 2006 for each Defining parameter
# Air Quality Index (AQI)
import pandas as pd
import datetime as datetime
dfAQI = pd.read_csv("https://aqs.epa.gov/aqsweb/airdata/daily_aqi_by_cbsa_2006.zip", sep = ',')
dfAQI = dfAQI[['CBSA', 'Date', 'Category', 'AQI', 'Defining Parameter']]
dfAQI['Date'] = pd.to_datetime(dfAQI.Date)
dfAQI['month_year']=dfAQI.Date.dt.to_period('M')
dfAQI = dfAQI.groupby(['month_year','Defining Parameter']).AQI.mean()
dfAQI
```

```python
# mean of monthly AQI  grouped by Defining Parameter for the year 2006
# sorted by mean in descending order
import pandas as pd
import datetime as datetime
dfAQI = pd.read_csv("https://aqs.epa.gov/aqsweb/airdata/daily_aqi_by_cbsa_2006.zip", sep= ',')
dfAQI['Date'] = pd.to_datetime(dfAQI.Date)
dfAQI['month_year']=dfAQI.Date.dt.to_period('M')
dfAQI = dfAQI.groupby(['month_year','Defining Parameter']).AQI.agg(['mean', 'max'])
dfAQI =  (dfAQI.sort_values(by='mean', ascending=False))
dfAQI.head(10)
```
