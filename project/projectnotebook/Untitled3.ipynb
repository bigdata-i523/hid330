```python
import pandas as pd
import datetime as datetime
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
dfAQI = pd.read_csv("Desktop/Project/daily_aqi_by_cbsa_2016.csv", sep= ',')
dfAQI['Date'] = pd.to_datetime(dfAQI.Date)
dfAQI = dfAQI[['CBSA', 'Date', 'AQI', 'Category', 'Defining Parameter','Number of Sites Reporting']]
dfAQI['Defining Parameter'].value_counts()
sns.FacetGrid(dfAQI, hue ='Defining Parameter', size=10)\
    .map(plt.scatter, 'Number of Sites Reporting', 'AQI')\
    .add_legend()

```

```python
# Box Plot For No of days of each Category
import pandas as pd
import datetime as datetime
import matplotlib.pyplot as plt
%matplotlib inline
dfAQI = pd.read_csv("Desktop/Project/daily_aqi_by_cbsa_2016.csv", sep= ',')
dfAQI['Date'] = pd.to_datetime(dfAQI.Date)
dfAQI = dfAQI[['CBSA', 'Date', 'Category', 'Defining Parameter']]
pldata = dfAQI.groupby(['CBSA','Category','Defining Parameter']).count()
pldata.boxplot(column = 'Date',by='Category', manage_xticks =True, figsize =(15, 10), patch_artist = True)
```
